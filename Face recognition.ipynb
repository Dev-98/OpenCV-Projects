```python
import numpy as np
import cv2


from sklearn.svm import SVC

df = np.load('all_face_data.npy')

# print(df.shape,df.dtype)
x = df[:,1:].astype(int)
y = df[:,0]

knn = SVC().fit(x,y)

video = cv2.VideoCapture(0)

detector = cv2.CascadeClassifier('C:\\Users\\Dev Gupta\\Documents\\python\\Face recognization\\haarcascade_frontalface_default.xml')


while True:

    ret, frame = video.read()
    frame = cv2.flip(frame,1)
    
    if ret:

        faces = detector.detectMultiScale(frame)

        for face in faces:

            x, y, w, h = face
            
            cut = frame[y:y+h , x:x+w]
            fix = cv2.resize(cut,(200,200))
            gray = cv2.cvtColor(fix,cv2.COLOR_BGR2GRAY)

            out = knn.predict([gray.flatten()])

            cv2.rectangle(frame,(x,y),(x+w, y+h),(245,0,235),3)
            cv2.putText(frame,str(out[0]),(x,y-10),cv2.FONT_ITALIC,2,(0,255,0),2)


#             cv2.imshow('My face',gray)
        
        cv2.imshow('My screen',frame)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    # if key == ord('c'):


video.release()
cv2.destroyAllWindows()

```

# Face mood recogn 

```python
import numpy as np
import cv2
import dlib
# from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

df = np.load('face_mood_data2.npy')

x = df[:,1:].astype(int)
y = df[:,0]

knn = SVC().fit(x,y)

video = cv2.VideoCapture(0)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('C:\\Users\\Dev Gupta\\Documents\\python\\Face recognization\\shape_predictor_68_face_landmarks.dat')

while True :
    
    ret , frame = video.read()
    # frame = cv2.flip(frame,1)
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 

    faces = detector(gray)
    
    for face in faces:
        
        landmark = predictor(gray,face)
        
        expression = np.array([[point.x , point.y] for point in landmark.parts()[17:]])
        
        out = knn.predict([expression.flatten()])
        
        point = landmark.parts()[19]
        cv2.putText(frame,str(out[0]),(point.x,point.y),cv2.FONT_ITALIC,2,(130,255,195),2)
        
    if ret:
        
        cv2.imshow('yoyoy',frame)
        
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
        

video.release()
cv2.destroyAllWindows()
```
